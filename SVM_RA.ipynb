```python
#Function for extracting features from the different data frames
#Features are: Mean, STD, Skewness, Kurtosis, Variance, IQR
def features(dataframe):
  count = 0
  #Store mean values
  x_mean = []
  y_mean = []
  z_mean = []
  #Store standard deviation values
  x_std = []
  y_std = []
  z_std = []
  #Store max
  x_max = []
  y_max = []
  z_max = []
  #Store min
  x_min = []
  y_min = []
  z_min = []

  #Store Median
  x_med = []
  y_med = []
  z_med = []
  #List of all features extracted
  feature_extraction = []


  for i in range(len(dataframe)):
    if count != 99:
      #Append to mean list
      x_mean.append(dataframe['x'][i])
      y_mean.append(dataframe['y'][i])
      z_mean.append(dataframe['z'][i])

      #Append to standard deviation list
      x_std.append(dataframe['x'][i])
      y_std.append(dataframe['y'][i])
      z_std.append(dataframe['z'][i])

      #Append to max
      x_max.append(dataframe['x'][i])
      y_max.append(dataframe['y'][i])
      z_max.append(dataframe['z'][i])

      #Append to min
      x_min.append(dataframe['x'][i])
      y_min.append(dataframe['y'][i])
      z_min.append(dataframe['z'][i])

      #Append to Median
      x_med.append(dataframe['x'][i])
      y_med.append(dataframe['y'][i])
      z_med.append(dataframe['z'][i])


      count = count + 1
    else:
      #Calculate mean
      x_mean = mean(x_mean)
      y_mean = mean(y_mean)
      z_mean = mean(z_mean)

      #Calculate standar deviation
      x_std = stdev(x_std)
      y_std = stdev(y_std)
      z_std = stdev(z_std)

      #Calculate max
      x_max = max(x_max)
      y_max = max(y_max)
      z_max = max(z_max)

      #Calculate min
      x_min = min(x_min)
      y_min = min(y_min)
      z_min = min(z_min)

      #Calculate median
      x_med = median(x_med)
      y_med = median(y_med)
      z_med = median(z_med)

      #Store features in tuple then store tuple in feature_extracion list
      data_tuple = (x_mean, y_mean, z_mean, x_std, y_std, z_std, x_max, y_max, z_max, x_min, y_min, z_min, x_med, y_med, z_med)
      feature_extraction.append(data_tuple)

      #Clear lists to proceed to the next 5 elements
      x_mean = []
      y_mean = []
      z_mean = []

      x_std = []
      y_std = []
      z_std = []

      x_max = []
      y_max = []
      z_max = []

      x_min = []
      y_min = []
      z_min = []

      x_med = []
      y_med = []
      z_med = []

      #Reset counter
      count = 0

  return feature_extraction #returns tuple list
```
